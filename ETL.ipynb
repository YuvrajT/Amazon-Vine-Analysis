```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Cloud ETL").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
# Read in Data from S3 Buckets

from pyspark import SparkFiles

url= "https://classmodule.s3.us-east-2.amazonaws.com/user_data.csv"

spark.sparkContext.addFile(url)

user_data_df = spark.read.csv(SparkFiles.get("user_data.csv"), sep=",", header=True, inferSchema=True)
```

```python
# Show Data Frame

user_data_df.show()
```

```python
url= "https://classmodule.s3.us-east-2.amazonaws.com/user_payment.csv"

spark.sparkContext.addFile(url)

user_payment_df = spark.read.csv(SparkFiles.get("user_payment.csv"), sep=",", header=True, inferSchema=True)

user_payment_df.show()
```

Transform

```python
#join the two Data Frames
joined_df = user_data_df.join(user_payment_df, on="username", how="inner")
joined_df.show(5)
```

```python
# Droping null values

dropna_df = joined_df.dropna()
dropna_df.show(5)
```

```python
#load in Sql function to use columns 

from pyspark.sql.functions import col

#filter for Active User

cleaned_df = dropna_df.filter(col("active_user")==True)
cleaned_df.show()
```

```python
#Create user Data Frame to match active user table 

clean_user_df = cleaned_df.select(["id", "first_name", "last_name","username"])
clean_user_df.show()
```

```python
#create a data frame to match billin info table 

clean_billing_df = cleaned_df.select(["billing_id","street_address","state", "username"])
```

```python
#create a data frame to match payment info

clean_payment_df = cleaned_df.select(["billing_id","cc_encrypted"])
```
